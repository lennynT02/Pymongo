```python
'''
GRUPO 3
Lenin Masapaanta
Anthony Plua
Erick Villarroel
'''
import pandas as pd
from pymongo import MongoClient
client = MongoClient('127.0.0.1', 27017)
print(client.list_database_names())
database=client.Pokemon #Selecionamos la base de datos
print(database.list_collection_names()) #Mostramos las colecciones de la base de datos
collection = database['pokemon']
for i in collection.find([],{'_id':0}).sort('name',-1):
    print(i)
```

```python
# Crear un documento (insertar)
#,Name,Type 1,Type 2,Total,HP,Attack,Defense,Sp. Atk,Sp. Def,Speed,Generation,Legendary
documento1 = {"#": "722", "Name": "hola", "Type 1": "Fyre", "Type 2":"Rock", "Total": "600", "HP": "80", "Attack": "120", "Defense": "80", "Sp. Atk": "120", "Sp. Def": "80", "Speed": "120", "Generation": "6", "Legendary": "True"}
collection.insert_one(documento1)
info=collection.find_one({"Name":"hola"})
print(info) #Mostramos el documento creado
```

```python
# Leer documentos (consultar)
consulta1 = collection.find({"Type 1": "Rock"}).limit(5)
for documento in consulta1:
    print(documento)
```

```python
# Actualizar un documento
filtro = {"Name": "hola"}
nuevos_datos = {"$set": {"Name": "HOLA"}}
collection.update_one(filtro, nuevos_datos)
info=collection.find_one({"Name":"HOLA"})
print(info) 
```

```python
# Borrar un documento
filtro_borrar = {"Name": "HOLA"}
collection.delete_one(filtro_borrar)
info=collection.find_one({"Name":"hola"})
print(info)
```

```python
#1CONSULTA
consulta_regex = collection.find({"Type 2": {"$regex": "^Rock"}}).limit(5)
for documento in consulta_regex:
    print(documento)
```

```python
#2CONSULTA 
import pymongo
consulta_ordenada = collection.find().sort("HP", pymongo.ASCENDING).limit(5)
for documento in consulta_ordenada:
    print(documento)
```

```python
#3CONSULTA 
consulta_logica = collection.find({"$or": [{"Attack": 90}, {"Defense": 230}]}).limit(5)
for documento in consulta_logica:
    print(documento)
```

```python
#4CONSULTA 
consulta_proyeccion = collection.find({"Type 1": "Grass"}, {"Attack": 40, "_id": 0}).limit(5)
for documento in consulta_proyeccion:
    print(documento)
```

```python
#5CONSULTA
consulta_orden_desc =collection.find().sort("#", pymongo.DESCENDING).limit(5)
for documento in consulta_orden_desc:
    print(documento)
```

```python
#6CONSULTA
consulta_in = collection.find({"Generation": {"$in": [5, 9]}}).limit(5)
for documento in consulta_in:
    print(documento)
```

```python
#7CONSULTA
consulta_hp_superior_80 =collection.find({"HP": {"$gt": 80}}).limit(5)
for pokemon in consulta_hp_superior_80:
    print(pokemon)
```

```python
#8CONSULTA
consultaLegendarios = collection.find({"Legendary": True}).limit(5)
for pokemon in consultaLegendarios:
    print(pokemon)
```

```python
#9CONSULTA
consulta_velocidad_superior_50 = collection.find({"Speed": {"$gt": 50}}).sort("Speed", pymongo.DESCENDING).limit(5)
for pokemon in consulta_velocidad_superior_50:
    print(pokemon)
```
